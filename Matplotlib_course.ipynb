By default, the plot() function draws a line **from point to point.**

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,6])
ypoints=np.array([3,4])
plt.plot(xpoints,ypoints)
plt.show()
```

### Plotting Without Line
use shortcut string notation parameter 'o'

```python
plt.plot(xpoints,ypoints,'o')
plt.show()
```

### Multiple Points
Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

### Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.


```python
ypoints=np.array([5,8,3,1])
plt.plot(ypoints)
plt.show()
```

## Markers


```python
plt.plot(ypoints,marker='v')
plt.show()
```

sinon on peut utiliser les paramétres :

**mec** :  set the color of the edge of the markers

**mfc** :  set the color inside the edge of the markers

**ms** :  set the size of the markers

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

### Matplotlib Line
**linestyle** or **ls** : change the style of the plotted line

**color** or **c** : set the color of the line

**linewidth** or **lw** : change the width of the line



```python
plt.plot(ypoints,linestyle='-' , color='r' , linewidth = '10.5')
plt.show()
```

## Matplotlib Labels and Title

### Create Labels and a title for a Plot


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Set Font Properties for Title and Labels


```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

### Position the Title
loc : position the title.

values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
plt.title("Sports Watch Data", loc = 'left')

plt.plot(x,y)
plt.show()
```

## 2. Matplotlib Scatter
**scatter** : plots one dot for each observation.chaque point est dessiné indépendamment, sans ligne.

**c** : colors

**cmap** : color map

**s**: sizes

**alpha** : transparency


```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()

#rq : plot() relie par défaut les points par des lignes,mais le 'o' fait un "détournement" de plot pour faire un scatter.
```

### Colors

```python
plt.scatter(x, y, color = 'hotpink')
plt.show()

#Color Each Dot
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

### ColorMap
The Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

for exemple viridis' is one of the built-in colormaps available in Matplotlib, it ranges from 0, which is a purple color, up to 100, which is a yellow color.

**In addition you have to create an array with values (from 0 to 100), one value for each point in the scatter plot:**


```python
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

You can include the colormap in the drawing by including the plt.colorbar() :

```python
plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

### Size of each dot
You can change the size of the dots with the **s** argument.



```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

```python
#tous les points meme size
plt.scatter(x, y, s=50)

plt.show()
```

### Alpha  : 
adjust the **transparency** of the dots

```python
plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

## Matplotlib Pie Charts


```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 

#the pie chart draws one piece (called a wedge) for each value in the array(35 is the biggest 
# one(blue) , 25 is the orange...)
```



RQ: The size of each wedge is determined by comparing the value with all the other values, by using this formula:

                 The value divided by the sum of all values: x/sum(x)

### Labels

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

### Start Angle



```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

## Explode
Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and **not None**, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

### Shadow

```python
y = np.array([35, 25, 25, 15])

plt.pie(y,  shadow = True)
plt.show() 
```

### Colors

```python
y = np.array([35, 25, 25, 15])
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, colors = mycolors)
plt.show() 
```

### Legend 
(with the labels)

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:", loc="upper right")
plt.show() 
```
