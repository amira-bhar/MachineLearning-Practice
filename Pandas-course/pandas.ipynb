## Indexing in pandas


Pandas has its own accessor operators, **loc** and **iloc**

**Index-based selection : iloc** : selecting data based on its *numerical* position in the data.

           df.iloc[row position][column position]

```python
#read one row
df.iloc[0]
#read one column
df.iloc[:,1]
#read some rows of a column
df.iloc[0:3,1]
#It's also possible to pass a list:
df.iloc[[0,1,2],1]
#last five rows
df.iloc[-5:]
```

**Label-based selection : loc** :  it's the data index value, not its position, which matters.

           df.loc[index value][header(column) value]

 <span style="color:red">
 
 **RQ1:**

  df.iloc[] → par **position** (0 = première ligne, 1 = deuxième ligne, …)

df.loc[] → par **étiquette d’index** (la valeur de l’index, qui peut être 0, 1, 2… ou tout autre nombre/texte)</span>

**RQ2:**

iloc -> 0:10 will select entries 0,...,9.

loc  -> So 0:10 will select entries 0,...,10.


```python
df.loc[1:3,'Name']
df.loc[:, ['Name', 'Type 1']]
```

### Maps

**.map()** permet d’appliquer une opération élément par élément sur une Series (une seule colonne).


```python
HP_mean=df['HP'].mean()
#remean HP values to 0
df['HP'].map(lambda x:x-HP_mean)
```

**apply()** is the equivalent method if we want to transform a whole DataFrame by calling a custom method on each row.



```python
#changer la colonne HP
def remean_points(row):
    row.HP = row.HP - HP_mean
    return row

df.apply(remean_points, axis='columns') #axis='columns' (ou axis=1), il va parcourir chaque ligne une par une
#row est une Serie représentant une ligne.
#à chaque itération, il transforme la ligne en un petit pandas.Series
```

voici un exemple d'une ligne (row) dans une itération  : 

| index | wine | points |
| ----- | ---- | ------ |
| 0     | A    | 90     |

la ligne(row) est transformée en une Serie (après passage par .apply(axis=1)) :

**wine**         A        
**points**      90        
Name:        0             

**Les noms de colonnes** (wine, points) deviennent l’**index** de la Series.

Les valeurs de la ligne ("A", 90) deviennent les valeurs de la Series.

Le numéro de la ligne originale (0) est affiché comme le name de la Series.

row.column_name <=> row.index

**RQ** : 

axis=0 ou "index" → ça parcourt chaque colonne (fonction appliquée colonne par colonne).

axis=1 ou "columns" → ça parcourt chaque ligne (fonction appliquée ligne par ligne).



Note that map() and apply() return new, transformed **Series and DataFrames, respectively.**

**They don't modify the original data they're called on.** If we look at the first row of df, we can see that it still has its original HP value.

apply n’a pas le même comportement selon qu’on l’utilise sur un DataFrame ou sur un GroupBy !!!!!

1️⃣ DataFrame.apply()

-Quand tu appelles apply directement sur un DataFrame :

df.apply(fonction)


Par défaut axis=0 → donc fonction est appliquée colonne par colonne

Si tu mets axis=1 → fonction est appliquée ligne par ligne

👉 Ici, la fonction reçoit une Series (pas le DataFrame entier) :

axis=0 → chaque colonne devient une Series

axis=1 → chaque ligne devient une Series

2️⃣ GroupBy.apply()

df.groupby("col").apply(fonction)

La fonction reçoit un DataFrame entier correspondant à chaque groupe

Tu n’as pas besoin de mettre axis, car ça s’applique par groupe, pas par ligne/colonne

```python
import pandas as pd
df = pd.DataFrame({
    "group": ["A", "A", "B", "B"],
    "value": [10, 20, 30, 40]
})
print(df)
print(df.groupby("group").apply(lambda g: g["value"].mean()))
# 15=mean(10,20) et 35=mean(30,40)
```

you can also group by more than one column.

you can also run a bunch of different functions on your sub-DataFrame simultaneously. 

```python
df.groupby(["HP"]).HP.agg([len,max,min])
```

### Combining
*  concat()

```python
df1 = pd.DataFrame({"A": [1,2], "B": [3,4]})
df2 = pd.DataFrame({"A": [5,6], "B": [7,8]})
print(df1)
print(df2)
```

```python
print(pd.concat([df1, df2], axis=0))
```

```python
print(pd.concat([df1, df2], axis=1))
```

* join() : Join **par index**

```python
df1 = pd.DataFrame({"A": [1,2]}, index=["x","y"])
df2 = pd.DataFrame({"B": [3,4]}, index=["x","z"])
print(df1)
print(df2)
```

```python

print(df1.join(df2, how="left")) #how="left" garde l’index de df1 , how="right" garde l'index de df2
```

```python
print(df1.join(df2, how="outer")) #union des deux indexes
```

```python
#si les deux dfs ont meme header
df1 = pd.DataFrame({"A": [1, 2]}, index=["x", "y"])
df2 = pd.DataFrame({"A": [3, 4]}, index=["x", "y"])
print(df1.join(df2, lsuffix="_left", rsuffix="_right"))
#Si les colonnes ont le même nom → on doit mettre des suffixes pour éviter le conflit.
```
